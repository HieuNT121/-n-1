```python
import pandas as pd
import folium
from folium.plugins import BeautifyIcon
```

```python

## Import the geolocation data of Quan Hoa, Cầu Giấy
data = pd.read_csv('VN2000 - WGS84.csv')
```

```python
lon = data['WGS84 - Long']
lat = data['WGS84 - Lat']

SGmap = folium.Map(location=[21.034870, 105.795905], tiles="OpenStreetMap", zoom_start=35) 
for i in range(len(lon)):
    folium.Marker(
        location=[lat[i], lon[i]],
        icon=BeautifyIcon(
            icon_shape='marker',
            number=int(i + 1),
            spin=True,
            text_color='red',
            background_color="#FFF",
            inner_icon_style="font-size:12px;padding-top:-5px;"
        )
    ).add_to(SGmap)
SGmap
```
